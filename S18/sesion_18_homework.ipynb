```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

plt.plot(df["Age"], df["Sleep Duration"], label="Sleep Duration")
plt.xlabel("Age")
plt.ylabel("Hours of Sleep")
plt.legend()
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Create a scatter plot of sleep duration vs. stress level
fig = px.scatter(df, x="Sleep Duration", y="Stress Level", color="Gender")
fig.show()
```
