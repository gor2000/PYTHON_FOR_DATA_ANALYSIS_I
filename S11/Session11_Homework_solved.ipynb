### Exercise 4
What's the best selling `product_id`? Use `for` loops to find out.

You can loop through a dataframe using `iterrows()` like this:

```python
for index, row in df.iterrows():
    print(index, row)
```

where `index` is the row index and `row` is a series of the row values.

````python
product_ids = ecomm["product_id"].unique()

results = {}

for product_id in product_ids:

    total_units_sold = 0

    for idx, row in ecomm.iterrows():

        if row["product_id"] == product_id:
            total_units_sold += row["units"]

    results[product_id] = total_units_sold

#Â sort the dictionary by value
sorted_results = sorted(results.items(), key=lambda x: x[1], reverse=True)
````

````python
sorted_results
````

Now check how easy it is to calculate the same thing using groupby

````python
# calculate best selling product using groupby

ecomm.groupby("product_id").sum().sort_values("units", ascending=False)
````
